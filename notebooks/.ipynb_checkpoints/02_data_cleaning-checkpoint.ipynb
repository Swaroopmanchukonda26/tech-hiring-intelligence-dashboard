```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load all 3 datasets
layoffs = pd.read_csv('../data/raw/layoffs/layoffs.csv')
placement = pd.read_csv('../data/raw/dataset 2/student_placement_synthetic.csv')
hr = pd.read_csv('../data/raw/IBM dataset/WA_Fn-UseC_-HR-Employee-Attrition.csv')

print("✅ All datasets loaded!")
```

```python
print("🔴 LAYOFFS — Missing Values:")
print(layoffs.isnull().sum())
print(f"\nTotal missing: {layoffs.isnull().sum().sum()}")
```

```python
print("🟡 PLACEMENT — Missing Values:")
print(placement.isnull().sum())
print(f"\nTotal missing: {placement.isnull().sum().sum()}")
```

```python
print("🟢 IBM HR — Missing Values:")
print(hr.isnull().sum())
print(f"\nTotal missing: {hr.isnull().sum().sum()}")
```

```python
# Make a copy first — always work on copy!
layoffs_clean = layoffs.copy()

# Drop rows where company name is missing
layoffs_clean = layoffs_clean.dropna(subset=['company'])

# Fill missing total_laid_off with 0
layoffs_clean['total_laid_off'] = layoffs_clean['total_laid_off'].fillna(0)

# Fill missing industry with 'Unknown'
layoffs_clean['industry'] = layoffs_clean['industry'].fillna('Unknown')

# Fill missing country with 'Unknown'
layoffs_clean['country'] = layoffs_clean['country'].fillna('Unknown')

# Convert date column to datetime format
layoffs_clean['date'] = pd.to_datetime(layoffs_clean['date'])

# Remove duplicate rows
layoffs_clean = layoffs_clean.drop_duplicates()

print("✅ Layoffs cleaning done!")
print(f"Before: {layoffs.shape[0]} rows")
print(f"After: {layoffs_clean.shape[0]} rows")
```

```python
# Make a copy
placement_clean = placement.copy()

# Check and remove duplicates
placement_clean = placement_clean.drop_duplicates()

# Fill any missing values with median
for col in placement_clean.select_dtypes(include=[np.number]).columns:
    placement_clean[col] = placement_clean[col].fillna(placement_clean[col].median())

print("✅ Placement cleaning done!")
print(f"Before: {placement.shape[0]} rows")
print(f"After: {placement_clean.shape[0]} rows")
```

```python
# Make a copy
hr_clean = hr.copy()

# Remove duplicates
hr_clean = hr_clean.drop_duplicates()

# Check for missing values
print("Missing values in HR data:", hr_clean.isnull().sum().sum())

print("✅ IBM HR cleaning done!")
print(f"Before: {hr.shape[0]} rows")
print(f"After: {hr_clean.shape[0]} rows")
```

```python
# Save all cleaned datasets
layoffs_clean.to_csv('../data/cleaned/layoffs_clean.csv', index=False)
placement_clean.to_csv('../data/cleaned/placement_clean.csv', index=False)
hr_clean.to_csv('../data/cleaned/hr_clean.csv', index=False)

print("✅ All cleaned datasets saved!")
print("📁 Location: data/cleaned/ folder")
print(f"\n🔴 Layoffs clean: {layoffs_clean.shape}")
print(f"🟡 Placement clean: {placement_clean.shape}")
print(f"🟢 HR clean: {hr_clean.shape}")
```
